```python
import matplotlib.pyplot as plt
import numpy as np


xpoints = np.array([0.0,1.0,2.0,3.0])
ypoints = np.array([3,8,1.5,10])

plt.plot(xpoints,ypoints,'o:r')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([95,87,81,96])
mylabels = ["cse", "mech", "civil", "ece"]
plt.pie(y, labels = mylabels)
plt.legend()
plt.show() 
```

```python
import matplotlib.pyplot as plt
import numpy as np


xpoints =  np.array(['9:10','11:10','13:10','15:10'])
ypoints =  np.array(['evng','python','physics','chemistry'])

plt.plot(xpoints,ypoints,'o:m')
plt.xlabel("hours")
plt.ylabel("subjects")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([95,87,81,96])
mylabels = ["python", "chemistry", "english", "maths"]
myexplode=[0.2,0,0,0]
plt.pie(y,labels=mylabels, explode=myexplode)
plt.show() 
```
