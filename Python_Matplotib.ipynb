```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([0, 60, 25, 30])
ypoints = np.array([0, 250, 320, 420])

plt.plot(xpoints, ypoints)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["APPLE", "BALL", "CAT", "DOG"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

y = np.array([35, 30, 25, 60])
a= ["math","english","physics","chemistry"]

plt.pie(y,labels=a)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

xpoints = np.array([1, 8, 10, 16])
ypoints = np.array([3, 10, 22, 44])

plt.plot(xpoints, ypoints, 'o')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x, y, color = "green")
plt.show()
```
